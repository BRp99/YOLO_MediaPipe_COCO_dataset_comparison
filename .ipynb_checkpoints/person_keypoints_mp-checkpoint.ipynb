# Load annotations from COCO dataset

```python
dataDir = './coco/'
dataType = 'val2017'
annFile = '{}/annotations/instances_{}.json'.format(dataDir, dataType)
coco = COCO(annFile)
```

# Display COCO categories and supercategories

```python
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories names: \n{}\n'.format(' '.join(nms)))
```

# Get category 'snowboard'

```python
catIds = coco.getCatIds(catNms=['snowboard'])
imgIds_snowboard = coco.getImgIds(catIds=catIds)[:10]
print(imgIds_snowboard)

for img_id in imgIds_snowboard:
    img_info = coco.loadImgs(img_id)[0]
    
    I = io.imread(img_info['coco_url'])
    plt.axis('off')
    plt.imshow(I)
    plt.show()

# Get first and last image of 'imgIds_snowboard'and store this two images inside folder 'extracted_images_from_coco' 

ten_images = []

first_image_info = coco.loadImgs(imgIds_snowboard[0])[0]
ten_images.append(first_image_info)


last_image_info = coco.loadImgs(imgIds_snowboard[-1])[0]
ten_images.append(last_image_info)

print(ten_images)
```

# Create 'extracted_images_from_coco' folder to store images without keypoints

```python
input_folder = "./extracted_images_from_coco"
if not os.path.exists(input_folder):
    os.makedirs(input_folder)
```

```python
for image_info in ten_images:
    image_url = image_info['coco_url']
    file_name = image_info['file_name']
    image_path = os.path.join(input_folder, file_name)
    
    response = requests.get(image_url)
    
    
    with open(image_path, 'wb') as f:
        f.write(response.content)
```
